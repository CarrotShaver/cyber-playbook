# Cisco Switch Configuration
1. Download switch driver
2. Make sure to have console cable connected to switch
3. Download PuTTY, run and open for serial port (find port using device manager and under ports)

Notes:
- Switch to router connection is with ethernet cable
- Router to router connection is with serial cable (parallel router configuration)

### PuTTY Switch Config Commands
- Enable configuring `>enable`
    - Check current configurations `#show running-config`
    - Note: using the `#do` command before any show command will force it to run even if not in the right mode
    - Enter configuration mode `#config t`
        - Exit configuration mode `#exit`
    - See hardware information: `#show version`
    - Check interfaces: `#show interface vlan1`
    - Check if a vlan is open to be configured: `#show ip interface vlan1`
        - If it is open it should say Vlan(x) is up
    - Check a port and see what is configured on it: `#show interface fastethernet 0/1`
        - shorter: `#show int f0/1`
    - View directory where config files are stored: `#show flash` or `#dir`
    - Change directories: `#cd flash:otherdir`
    - Switch to configuration mode `#config t`
        #### Security
        - Change hostname: `#hostname *name*`
        - Switch to the console cable line: `#line console 0`
            - Disable messages when typing commands: `logging synchronous`
            - Disable dns lookup: `#no ip domain-lookup`
            - Change password for the console: `#password ratdick`
            - Enable password logins: `#login`
            - Change/enable the password for enable mode: `#enable password ratdick`
            - Enter telnet connection for theoretical ports 0-4: `#line vty 0 4`
            - Set up a password for telnet: `#password ratdick`
            - Enable password logins: `#login`
            - Encrypts passwords: `#enable secret yourpasswordhere` (should be different from other passwords)
            - Further encrypts passwords: `#service password-encryption`
            - Unencrypts password encryption: `#no service password-encryption`
            - Set up motd: `#banner motd "Message"`
            - Exit back to enable mode: `#exit`
            - Save all configurations: `#copy run start`
            
### Creating a VLAN
- Plan out a topology using Cisco Packet Tracer or similar, plan ip's and vlans
- Set up static IP for ethernet ipv4 on devices
- Enter into config mode: `config t`
    - Close all unused ports: `#interface range fa0/1 - 24`
    - Shutdown those port ranges: `#shutdown`
    - Re-enable the ports being used: `#interface range fa0/1 - 4`
    - Toggle on mode access for ports: `#switchport mode access`
    - Revert the shutdown on these ports: `#no shutdown`
    - **Configure VLANs on the switch**
        - Create a vlan: `vlan xx` (sub xx for the vlan numnbers)
        - Name the vlan: `name vlanxx`
        - Check for created vlans: `show vlan brief`
        - Add a port to a vlan:
            - Switch to interface for correct port: `#interface range fa0/x`
            - Add the port to a vlan you created: `#switchport access vlan xx`
            - Exit out: `#end`
        - Save configurations: `#copy running-config startup-config` or `#copy run start` for short
        - Check a configured vlan: `show vlan id xx` (substitute vlan number for xx)
- Configure the management VLAN (be in config t mode)
    - Enter vlan 99's interface: `#interface vlan 99`
    - Set up the IP and subnet mask: `#ip address 192.168.xx.xx 255.255.255.0`
    - Turn on the settings: `#no shutdown`
- Configure the switch to switch connection:
    - Enter the port that the connected switch is using: `#interface range fa0/1`
    - Set up the switchport trunk: `#switchport mode trunk`
    - Specify the management VLAN for this port: `#switchport trunk native vlan 99`
    - Turn on the settings: `#no shutdown`
    - Exit configuration mode: `#end`
    - Check configurations: `#show interface trunk`
    - Test connections using ping: `#ping 192.168.xx.xx`
```
            
#### Reset Configuration Settings (if forgot password)
- Hold down mode button, unplug, plug back in, wait for text to appear on putty, release mode button
- Enter the following commands:
```
Switch: flash_init
Switch: dir flash:
Switch: rename flash:config.text flash:config.old
Switch: boot
```
- When asked if you would like to enter initial dialogue say 'no'
