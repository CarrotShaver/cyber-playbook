# Github

### Creating a Repo and linking to a directory
1. Log in to github and create/name repo
    - Note: Better to not create README in repo and create in your own directory instead
2. Open Git Bash or cmd
3. Initialize the local directory aws a git repository
```
$ git init
# Creates the local git repo
```

4. Add files using "git add" to the repository (gets files ready for a commit)
```
$ git add .
# Preps the repo to add changed files
```

5. Commit files to the repository
```
$ git commit -m "Commit Information"
# the -m "message" should be used to give information on what was changed for this commit
```

6. Go to the github repository and copy the remote repository URL
    - (ex. https://github.com/CarrotShaver/cyber-playbook.git)
7. Use the CLI to add the URL for where the repo will be located on github using the remote add command
```
$ git remote add origin "URL"
# Sets the new remote repo
$ git remote -v
# Command to verify the remote URL
```

8. Push the changes in the local repo to the GitHub repo
```
$ git push origin master
# Pushes the changes in the local repo up to the remote repository
```

- Note: This will not work correctly if a README was created in the github repo, use this command to override
```
$ git pull origin master --allow-unrelated-histories
# Merges the two repos and uses the flag to allow the differences
$ git push -u origin master
# -u is short for --set-upstream
```

### Undoing or "Checkout" a file in git
- Finds the differences between your version of the file vs the version saved on github
```
$ git diff "FILENAME"
```

- Reverts the file in your directory to the one saved on github
```
$ git checkout --"FILENAME"
```

### Updating a local repo to github
1. Open git bash or cmd
2. Do a git status to check what files have been changed in the local repo vs github repo
```
$ git status
```

3. Prepare changed items to the github repo with git add
```
$ git add .
```

4. Commit the changes and write message for changes
```
$ git commit -m "Message about changes"
```

5. Push the changes onto the github repo
```
$ git push
```

6. Check github again to make sure changes were pushed correctly
```
$ git status
```
